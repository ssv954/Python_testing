#### Step 6: Show dataframe by condition

```python
df.head()                                       # show top 5
df.head(20)                                     # show top 20

df.tail()                                       # show last 5
df.tail(20)                                     # show last 20

df[5:15]                                        # show by range

df.shape                                        # show dimension data

df.info                                         # show info data
# df.info()                                     # show info data

df.describe()                                   # show statistic of data
df[['Name', 'Sex', 'Age', 'PassengerId']]       # show specific column
```

```python
result = df[['Name', 'Sex', 'Age', 'PassengerId']]
result.drop(['Sex'], axis = 1)                          # drop specific column

df.iloc[0:3]                                            # choose data by range

df[df['Age'] > 60]                                      # show data by condition

df[(df['Age'] > 60) & (df['Sex'] == 'female')]          # show data by multi condition

df.drop([0, 1])                                         # drop data

df.drop([2, 3, 4]).head()                               # show top 5 after drop

# df.replace({'column_name': {'search': 'replace'}})    # replace data by column
df.replace({'Pclass': {1: '1st', 2: '2nd', 3: '3rd'}})

df.replace(0, 'Zero')                                   # replace whole data
df['Sex'].value_counts().sort_index(ascending=False)    # count data by column and sort condition
```

#### Step 7: Data Cleansing

```python
df = pd.read_csv('dataset/exercise.csv')

# df.dropna(inplace=True)                       # drop row of having NaN data

df.fillna(0, inplace=True)                      # replace row of having NaN data

df.drop([3], inplace=True)                      # drop specific row
# df.drop(~df['Sex'].str.contains('male'))

df.loc[2, 'Hight'] = '179.0'                    # replace column with specific index

df[df.duplicated()]                             # check duplication values

df.drop_duplicates(inplace=True)                # drop duplication values
df
```
