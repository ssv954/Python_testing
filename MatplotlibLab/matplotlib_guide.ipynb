#### Step 3: Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# by define value
x = np.array([2018,2019,2020,2021,2022])
y = np.array([502,1203,1999,2039,2254])

plt.bar(
        x, y, 
        color=['green', 'red', 'blue', 'yellow', 'orange'],
        edgecolor = 'lightgreen'
    )
plt.title('BCEL Gross')
plt.xlabel('Year')
plt.ylabel('Revenue (Million Kip)')
plt.show

# by url value
df = pd.read_csv('dataset/gdp.csv')
plt.bar(
        df['year'], df['gdp'], 
        color=['green', 'red', 'blue', 'yellow', 'orange'],
        edgecolor = 'lightgreen'
    )
plt.title('BCEL Gross')
plt.xlabel('Year')
plt.ylabel('Revenue (Million Kip)')
plt.show
```
